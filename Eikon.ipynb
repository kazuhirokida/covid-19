```python
ticker = pd.read_csv('ticker.csv')
```

```python
covid19 = ek.get_timeseries(ticker.Ticker.tolist())
```

```python
covid19.columns = covid19.columns.map(ticker.set_index('Ticker').Japanese.to_dict())
```

```python
jhu = pd.read_csv(
    'https://github.com/CSSEGISandData/COVID-19/raw/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv')
```

```python
jhu.head()
```

```python
US = jhu[jhu['Country/Region'] == 'US'].loc[:,'1/22/20':]
```

```python
US.columns = pd.to_datetime(US.columns,format='%m/%d/%y')
```

```python
US.index = ['米国']
```

```python
US = US.loc[:,str(covid19.index[0]):]
```

```python
covid19[['米国(WHO)']].T.append(US)
```

```python
covid19.drop(['米国(WHO)','中国(WHO)'],axis=1).T.append(US).tail()
```

```python
covid19.drop(['米国(WHO)','中国(WHO)'],axis=1).T.append(US).to_csv('covid19.csv')
```
